求一元二次方程 ax²+bx+c=0 的实解

```python
import math

a = float(input("a="))
b = float(input("b="))
c = float(input("c="))

d = b*b - 4*a*c

if d < 0:
    print("无实数解")
else:
    x1 = (-b + math.sqrt(d))/(2*a)
    x2 = (-b - math.sqrt(d))/(2*a)
    print("x1 =", x1)
    print("x2 =", x2)
```

统计字符串中字符类型

```python
s = input("输入字符串:")

a = 0
b = 0
c = 0
d = 0

for c in s:
    if c.isalpha():
        a += 1
    elif c.isdigit():
        b += 1
    elif c.isspace():
        c += 1
    else:
        d += 1

print("字母:",a)
print("数字:",b)
print("空格:",c)
print("其它:",d)
```

```python
class Add:

    def __init__(self,a,b):
        self.a = a
        self.b = b

    def result(self):
        return self.a + self.b

obj = Add(3,7)

print("结果:",obj.result())
```

求解线性方程组
\
3x−2y+z=−1
\
−x−y+z=−1
\
x+y+z=2

```python
import numpy as np

A = np.array([[3,-2,1],
              [-1,-1,1],
              [1,1,1]])

B = np.array([-1,-1,2])

# 方法1
sol1 = np.linalg.solve(A,B)

# 方法2
sol2 = np.linalg.inv(A).dot(B)

print("solve:",sol1)
print("inverse:",sol2)
```

面向对象画圆

```python
import numpy as np
import matplotlib.pyplot as plt

class Circle:

    def __init__(self,r):
        self.r = r

    def draw(self):

        t = np.linspace(0,2*np.pi,100)

        x = self.r*np.cos(t)
        y = self.r*np.sin(t)

        plt.plot(x,y)

        plt.axis("equal")

        plt.title("Circle")

        plt.show()

c = Circle(3)

c.draw()
```

直线+噪声点

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,10)

y = 2*x + 5

noise = y + np.random.normal(0,2,10)

p1,p2 = np.random.choice(range(10),2)

plt.scatter(x,noise)

plt.plot([x[p1],x[p2]],[noise[p1],noise[p2]])

plt.show()
```

求函数极值\
f(x)=x³+3x²+4

```python
import sympy as sp

x = sp.symbols('x')

f = x**3 + 3*x**2 + 4

df = sp.diff(f,x)

critical = sp.solve(df,x)

print("极值点:",critical)

for c in critical:
    print("f(",c,") =",f.subs(x,c))
```
