```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


# 自动寻找系统中的中文字体
font = matplotlib.font_manager.FontProperties(fname="/System/Library/Fonts/STHeiti Light.ttc")

plt.rcParams['font.family'] = font.get_name()
plt.rcParams['axes.unicode_minus'] = False
# 生成数据
x, y = make_moons(100, noise=0.5)

# 划分训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=0)

# =========================
# （1）K值对KNN影响
# =========================

k_range = range(1, 51)
accuracy_list = []

for k in k_range:
    model = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    model.fit(x_train, y_train)

    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)

    accuracy_list.append(acc)

# 画折线图
plt.figure()
plt.plot(k_range, accuracy_list)

plt.xlabel("K值")
plt.ylabel("预测准确率")
plt.title("K值对K近邻算法预测准确率的影响")

plt.show()


# =========================
# （2）距离类型影响
# =========================

k_values = [5,10,20]

euclidean_acc = []
manhattan_acc = []

for k in k_values:

    # 欧式距离
    model_e = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    model_e.fit(x_train, y_train)
    y_pred_e = model_e.predict(x_test)
    euclidean_acc.append(accuracy_score(y_test, y_pred_e))

    # 曼哈顿距离
    model_m = KNeighborsClassifier(n_neighbors=k, metric='manhattan')
    model_m.fit(x_train, y_train)
    y_pred_m = model_m.predict(x_test)
    manhattan_acc.append(accuracy_score(y_test, y_pred_m))


# 画柱状图
x_pos = np.arange(len(k_values))

plt.figure()

plt.bar(x_pos-0.2, euclidean_acc, width=0.4, label='欧式距离')
plt.bar(x_pos+0.2, manhattan_acc, width=0.4, label='曼哈顿距离')

plt.xticks(x_pos, k_values)

plt.xlabel("K值")
plt.ylabel("预测准确率")
plt.title("不同距离度量方式对KNN分类效果的影响")

plt.legend()

plt.show()
```
