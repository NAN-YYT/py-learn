文件读写

```python
from pathlib import Path

# 相对路径：当前脚本/Notebook 目录下的 data/test.txt
file_path = Path("data/test.txt")

# 如果父目录不存在，就自动创建
file_path.parent.mkdir(parents=True, exist_ok=True)

# 写入文件（不存在会创建，存在会覆盖）
with open(file_path, "w", encoding="utf-8") as f:
    f.write("hello")

# 读取文件
with open(file_path, "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
```

面向对象基础

```python
# 定义一个 Person 类
class Person:
    # 构造方法：创建对象时自动调用
    def __init__(self, name1):
        # 把传入的 name 保存为对象属性
        self.name = name1

    # 实例方法：对象可以调用的行为
    def greet(self):
        # 使用 f-string 输出自我介绍
        print(f"你好，我是{self.name}")

# 创建 Person 类的实例（对象）
p = Person("Alice")

# 调用对象方法
p.greet()
```
